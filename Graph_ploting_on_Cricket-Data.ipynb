```python
import seaborn as sns
import pandas as pd
import matplotlib as mpb
import matplotlib.pyplot as plt
```

```python
odi = pd.read_csv("ICC ODI Rankings.csv")
odi.head()
```

```python
wodi = pd.read_csv("ICC Women's ODI Team Rankings.csv")
wodi.head()
```

```python
def trim(x):
    a = x[:3] 
    return a
```

```python
odi["Team"] = odi["Team"].apply(trim)
odi.head()
```

```python
wodi["Team"] = wodi["Team"].apply(trim)
wodi.head()
```

```python
mpb.rcParams['font.size'] = 25
```

## Team Vs Point [ Men ODI ]

```python
#Team Vs Point
f,ax1 = plt.subplots(figsize = (20,10),)
ax = sns.barplot('Team','Points',data=odi)
```

## Team vs Rating [ Men ODI ] 

```python
#Team vs Rating
f,ax = plt.subplots(figsize = (20,10))
ax = sns.barplot('Team','Rating',data=odi)
```

## Team vs Rating  horizontal graph[ Men ODI ] 

```python
#Team vs Rating [horizontal]
f,ax = plt.subplots(figsize = (20,10))
ax = sns.barplot('Rating','Team',data=odi)
```

## match vs team [ Women ODI ]

```python
#match vs team
f,ax = plt.subplots(figsize = (20,10))
ax = sns.barplot('Matches','Team',data=wodi)
```

```python
wodi['gender'] = "female"
odi["gender"] = 'male'
```

```python
comb = pd.concat([wodi,odi],ignore_index=True)
comb
```

## Team vs points on combitations dataframe 

```python
#Team vs points on combitations dataframe 
f,ax = plt.subplots(figsize = (25,12))
ax = sns.barplot('Team','Points',data=comb,hue = "gender")
```

## Team Vs Match Scatter_Graph

```python
f,ax = plt.subplots(figsize = (25,12))
ax = sns.scatterplot('Team','Matches',data = comb,hue = "gender")
```

## Team Vs Match Line_Graph

```python
f,ax = plt.subplots(figsize = (25,12))
ax = sns.lineplot('Team','Matches',data = comb,hue = "gender")
```
